```python
from matplotlib import pyplot as plt
import numpy as np

width = 15
height = 13
seq1 = np.arange(width * height)

lines = []
for i,n in enumerate(seq1):
    row = i // width
    col = i % width
    lines.append((row, col))
    plt.text( row, col, f"{i}")

lines = np.array(lines)

plt.plot(lines[:,0], lines[:,1], marker="o")
```
